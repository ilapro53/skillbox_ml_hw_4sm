Обучите модель DBSCAN с параметром *min_samples=8*. Визуализируйте результат.

```python
import pandas as pd
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('clustering_hw.csv')
data = df[['x1', 'x2']]
data.head()
```

```python
dbscan_model = DBSCAN(eps=0.5, min_samples=8)
dbscan_model.fit(data);
```

```python
clusters_labels = dbscan_model.labels_
unique_clabels = np.unique(clusters_labels)

plt.figure(figsize=(10, 6))

color = [0, 0.1, 0, 1]
c_step = 1/(len(unique_clabels)+3)

for label in unique_clabels:

    color[1] += c_step
    plt.scatter(
        data['x1'][clusters_labels == label], 
        data['x2'][clusters_labels == label], 
        c=[color],
        label=f'Кластер {label}'
    )

plt.title('Результат кластеризации DBSCAN')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```
